```python
#Cleaning data
df = pd.read_csv("raw_data.csv")

df.isnull()
df.isna()

df.isnull().sum()

df.dropna()  #raw drop
df.dropna(axis = 1)  #col drop

df.fillna(0)

age_mean = df["age"].mean()

cleaned_data = df.copy()
age_mean = cleaned_data["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)

df.ffill()  #forward fill
df.bfill()  #backward fill
```

```python
#handle duplicated
df.duplicated()
df[["country", "gender"]].duplicated()
df2 = df.copy

df.drop_duplicates(inplace = True)
```

```python
#Data types
df.dtypes
df2 = df.copy()

df2 = df2.fillna(0)
df2["age"] = df2["age"].astype("int64").copy()
df2.dtypes

#Handle string value
df["gender"].str.capitalize()

df["name"].str.split(" ")

df["country"].str.contains("India" , case = False)
```

```python
#feature engineering

df2 = df.copy()

df2["tax"] = df2["income"].apply(lambda x : "20%" if x >= 50000 else "10%")

gender_map = {"Male" : "M", "Female" : "F", "Unknown" : "U"}
df2["gender"]  = df2["gender"].map(gender_map)

df2.assign(new_income = df2["income"] * 1.1)

df2["country"] = df2["country"].replace("USA", "United State")

df2.columns = ["Id", "Name", "Age", "Country", "Gender", "Income", "Tax"]
df2.rename(columns = {"Income":"Salary"})
df2.rename(index = {1: "First"})

df2.sort_values("Income")
df2.sort_values("Income", ascending = False)
sorted_df = df2.sort_values(["Income","Age"])

sorted_df.sort_index()

#reset
sorted_df.reset_index()
sorted_df.reset_index(drop = True)

#Ranking
sorted_df["Ranking"] = sorted_df["Income"].rank(ascending = False, method= "dense")
sorted_df

#Reorder
sorted_df[["Id", "Name", "Age", "Country",  "Income", "Tax","Gender"]]
```

```python
#grouping and aggregation
df.groupby("country")["income"].mean()
df.groupby("country")["income"].min()
df.groupby("country")["income"].max()

df.groupby("gender")["income"].mean()

df.groupby("country")["income"].agg(["mean", "max", "min"])
df.groupby("country")["income"].aggregate(["mean", "max", "min"])

df.groupby("country")["income"].agg(avg_salary="mean", max_salary="max", min_salary="min")

df.groupby("country").agg({
    "income" : "max",
    "age" : "mean"
})
df.groupby("country").agg(
    max_salary = ("income" , "max"),
    avg_age = ("age" , "mean")
)
```

```python
# Melt and pivot
df = pd.DataFrame({
    "country": ["USA", "USA", "India", "India"],
    "year": [2020, 2021, 2020, 2021],
    "sales": [100, 120, 90, 110],
    "profit": [20, 25, 18, 22],
    
})

melted_df = df.melt(
    id_vars = ["country", "year"],
    value_vars = ["sales", "profit"],
    var_name = "metrics",
    value_name = "value"
)

original = melted_df.pivot(
    index = ["country", "year"],
    columns = "metrics",
    values = "value"
)
```

```python
#Merging and joining data

df_customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Adam", "Bob", "Charlie", "Dave"]
})

df_orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104 ],
    "customer_id": [2, 1, 4, 5],
    "amount": [250, 120, 300, 180]
})

pd.merge(df_customers, df_orders, on = "customer_id")  #inner join
pd.merge(df_customers, df_orders, on = "customer_id", how = "left")  #left join
pd.merge(df_customers, df_orders, on = "customer_id" , how = "right")  #right join
pd.merge(df_customers, df_orders, on = "customer_id" , how = "outer")  #outer join
```
